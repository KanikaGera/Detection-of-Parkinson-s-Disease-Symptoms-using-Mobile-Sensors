```
from scipy import fftpack
from scipy.stats import skew
from scipy import stats
from scipy.stats import kurtosis
from sklearn.externals import joblib
import pandas as pd
import numpy as np
```

```
def mn(df,new_dataframe):
  X_features=['accX/s','accY/s','accZ/s']
  for i in X_features:
    df[i] = np.abs(fftpack.fft(df[i]))
  curr_time= df['total_seconds'].iloc[0]
  end_time= df['total_seconds'].iloc[-1]
  while(curr_time<=end_time):
    df3=df.loc[(df['total_seconds']>=curr_time) & (df['total_seconds']<=(curr_time+14))]
    X_mean= df3['accX/s'].mean()
    Y_mean= df3['accY/s'].mean()
    Z_mean= df3['accZ/s'].mean()
    X_mad= df3['accX/s'].mad()
    Y_mad= df3['accY/s'].mad()
    Z_mad= df3['accZ/s'].mad()
    X_std= np.nanstd(df3['accX/s'])
    Y_std= np.nanstd(df3['accY/s'])
    Z_std= np.nanstd(df3['accZ/s'])
    X_min= df3['accX/s'].min()
    Y_min= df3['accY/s'].min()
    Z_min= df3['accZ/s'].min()
    X_max= df3['accX/s'].max()
    Y_max= df3['accY/s'].max()
    Z_max= df3['accZ/s'].max()
    X_iqr= stats.iqr(df['accX/s'], interpolation = 'midpoint')
    Y_iqr= stats.iqr(df['accY/s'], interpolation = 'midpoint')
    Z_iqr= stats.iqr(df['accZ/s'], interpolation = 'midpoint')
    local_df = pd.DataFrame({"accX_mean": [X_mean],"accY_mean": [Y_mean],"accZ_mean":[Z_mean],"accX_mad": [X_mad],"accY_mad": [Y_mad],"accZ_mad":[Z_mad],"accX_std": [X_std],"accY_std": [Y_std],"accZ_std":[Z_std],"accX_min": [X_min],"accY_min": [Y_min],"accZ_min":[Z_min],"accX_max": [X_max],"accY_max": [Y_max],"accZ_max":[Z_max],"accX_iqr": [X_iqr],"accY_iqr": [Y_iqr],"accZ_iqr":[Z_iqr]})
    new_dataframe=new_dataframe.append(local_df,ignore_index = True)
    curr_time+=15
  return new_dataframe
```

```
def sma(data):
  return np.sum(data)/len(data)

def energy(data):
  return np.sum(data**2)/len(data)

def entropy(signal):
    eps=0.00000001
    s= energy(signal)
    entropy = -np.sum(s * np.log2(s + eps))
    return entropy

def sk(df,new_dataframe):
  curr_time= df['total_seconds'].iloc[0]
  end_time= df['total_seconds'].iloc[-1]
  while(curr_time<= end_time):
    df3=df.loc[(df['total_seconds']>=curr_time) & (df['total_seconds']<=(curr_time+14))]
    X_s = skew(df3['accX/s'])
    Y_s= skew(df3['accY/s'])
    Z_s= skew(df3['accZ/s'])
    X_kr= kurtosis(df3['accX/s'])
    Y_kr= kurtosis(df3['accY/s'])
    Z_kr= kurtosis(df3['accZ/s'])
    X_sma= sma(df3['accX/s'])
    Y_sma= sma(df3['accY/s'])
    Z_sma= sma(df3['accZ/s'])
    X_en= energy(df3['accX/s'])
    Y_en= energy(df3['accY/s'])
    Z_en= energy(df3['accZ/s'])
    X_entropy= entropy(df['accX/s'])
    Y_entropy= entropy(df['accY/s'])
    Z_entropy= entropy(df['accZ/s'])
    tremor= df3['Tremor/s'].iloc[0]
    local_df = pd.DataFrame({"accX_skew": [X_s],"accY_skew": [Y_s],"accZ_skew":[Z_s],"accX_kurtosis": [X_kr],"accY_kurtosis": [Y_kr],"accZ_kurtosis":[Z_kr],"accX_sma": [X_sma],"accY_sma": [Y_sma],"accZ_sma":[Z_sma],"accX_energy": [X_en],"accY_energy": [Y_en],"accZ_energy":[Z_en],"accX_entropy": [X_entropy],"accY_entropy": [Y_entropy],"accZ_entropy":[Z_entropy],"Tremor":[tremor]})
    new_dataframe=new_dataframe.append(local_df,ignore_index = True)
    curr_time+=15
  return new_dataframe
```
